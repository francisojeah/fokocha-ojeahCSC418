```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Function to load portrait images of class members
def load_images():
    images = {}
    image_dir = "../images/project-images/"
    for filename in os.listdir(image_dir):
        if filename.endswith(".jpg"):
            username = filename.split(".")[0].lower()
            images[username] = cv2.imread(os.path.join(image_dir, filename))
    return images

# Function for enhancing image using blue, green, red channel splitting
def enhance_color_channels(image):
    B, G, R = cv2.split(image)
    return B, G, R

# Function for enhancing image using weighted addition
def enhance_weighted_addition(image1, image2):
    #resize images
    image1 = cv2.resize(image1, (500,400))
    image2 = cv2.resize(image2, (500,400))
    return cv2.addWeighted(image1, 0.5, image2, 0.6, 0)

# Function for enhancing image using subtraction
def enhance_subtraction(image1, image2):
    #resize images
    image1 = cv2.resize(image1, (500,400))
    image2 = cv2.resize(image2, (500,400))
    return cv2.subtract(image1, image2)

# Function for enhancing image using sharpening
def enhance_sharpening(image):
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    return cv2.filter2D(image, -1, kernel)

# Function for enhancing image using median blur
def enhance_median_blur(image):
    return cv2.medianBlur(image, 15)

# Function for enhancing image by scaling
def enhance_scaling(image):
    return cv2.resize(image, None, fx=2, fy=2)

# Function for enhancing image by adjusting brightness and contrast
def enhance_brightness_contrast(image, brightness=5, contrast=1.5):
    return cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)

# Function for enhancing image by inverting colors
def enhance_inverse(image):
    return 255 - image
```

```python
# Load portrait images of class members
portrait_images = load_images()
```

```python
# Load the "superman.jpg" image
superman_image = cv2.imread("../images/project-images/superman.jpg")
```

```python
# Prompt for login
username = input("Enter your username (firstname): ").lower()
matno = int(input("Enter your matriculation number: "))

# Check if the username exists in the portrait images
if username not in portrait_images:
    print("No portrait found for the provided username.")
else:
    
    print(f"\nWelcome {username}!")
    
    # Menu for enhancement techniques
    print("\nAvailable enhancement techniques:")
    print("1. Color Channels Splitting")
    print("2. Weighted Addition")
    print("3. Subtraction")
    print("4. Sharpening")
    print("5. Median Blur")
    print("6. Scaling")
    print("7. Brightness & Contrast Adjustment")
    print("8. Inverse Color")

    choice = int(input("\nEnter your choice (1-8): "))

    # Apply selected enhancement technique to the portrait image
    original_image = portrait_images[username]
    print("\nEnhancing image...")
    enhanced_images = []
    titles = []

    if choice == 1:
        B, G, R = enhance_color_channels(original_image)
        enhanced_images = [B, G, R]
        titles = ["Blue Channel", "Green Channel", "Red Channel"]
    elif choice == 2:
        enhanced_images = [enhance_weighted_addition(original_image, superman_image)]
        titles = ["Weighted Addition with Superman"]
    elif choice == 3:
        enhanced_images = [enhance_subtraction(original_image, superman_image)]
        titles = ["Subtraction with Superman"]
    elif choice == 4:
        enhanced_images = [enhance_sharpening(original_image)]
        titles = ["Sharpened Image"]
    elif choice == 5:
        enhanced_images = [enhance_median_blur(original_image)]
        titles = ["Median Blurred Image"]
    elif choice == 6:
        enhanced_images = [enhance_scaling(original_image)]
        titles = ["Scaled Image"]
    elif choice == 7:
        enhanced_images = [enhance_brightness_contrast(original_image)]
        titles = ["Brightness & Contrast Adjusted Image"]
    elif choice == 8:
        enhanced_images = [enhance_inverse(original_image)]
        titles = ["Inverse Color Image"]
    else:
        print("Invalid choice.")

    # Display images
    if enhanced_images:
        plt.figure(figsize=(12, 6))

        # Display original image
        plt.subplot(1, len(enhanced_images) + 1, 1)
        plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
        plt.title("Original Image")
        plt.axis("off")

        # Display enhanced images
        for i, (enhanced_image, title) in enumerate(zip(enhanced_images, titles), start=2):
            plt.subplot(1, len(enhanced_images) + 1, i)
            if choice == 1:  # Color Channels Splitting
                plt.imshow(enhanced_image, cmap="gray")
            else:
                plt.imshow(cv2.cvtColor(enhanced_image, cv2.COLOR_BGR2RGB))
            plt.title(title)
            plt.axis("off")

        plt.tight_layout()
        plt.show()
```
