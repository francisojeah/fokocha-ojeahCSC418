# Robert Operator | OpenCV

```python
# Robert Edge Detection

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("../images/exercise-images/francis-profile.jpeg", cv2.IMREAD_GRAYSCALE)

# Perform Guassian Blur
image = cv2.GaussianBlur(image, (5, 5), 0)

# Define Robert kernels
kernel_x = np.array([[1, 0],
                     [0, -1]])

kernel_y = np.array([[0, 1],
                     [-1, 0]])

# Apply Robert edge detection
robert_x = cv2.filter2D(image, -1, kernel_x)
robert_y = cv2.filter2D(image, -1, kernel_y)

# Combine the gradient images
robert_combined = np.sqrt(np.square(robert_x) + np.square(robert_y))

# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original')
plt.axis('off')

# Display the Robert edges
plt.subplot(1, 2, 2)
plt.imshow(robert_combined, cmap='gray')
plt.title('Robert Edges')
plt.axis('off')

# Show the plot
plt.show()
```
