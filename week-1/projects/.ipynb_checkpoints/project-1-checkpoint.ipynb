```python
# Function to display profile based on username and password
def display_profile(username, password):
    
    matching_profiles = [profile for profile in profiles if profile["firstname"].lower() == username.lower()]
    
    if not matching_profiles:
        print("\nProfile not found for the provided username.")
        return
    
    for profile in matching_profiles:
        fullname = f"{profile['title']} {profile['firstname']} {profile['middlename']} {profile['lastname']}".strip()
        
        if sum(c.isalpha() for c in profile["lastname"]) == password:
            
            print(f"\nProfile found for username: {username}\n")
            
            # Load and display the corresponding image
            image = cv2.imread(profile["image_path"])
            plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
            plt.axis('off')
            plt.show()
            
            # Display profile information
            print(f"{fullname}\n")
            print(profile["description"])
            
            return
    
    print("Incorrect password.")
```

```python
# Request username and password
username = input("Enter your first name: ")
password = int(input("Enter your password (sum of alphabets in your surname): "))
    
# Display profile
display_profile(username, password)
```
