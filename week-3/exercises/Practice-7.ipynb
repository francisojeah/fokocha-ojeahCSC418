# Image Blurring

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread("../images/exercise-images/sst.jpg")

# Create a figure and set its size
plt.figure(figsize=(10, 8))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Gaussian Blur
Gaussian = cv2.GaussianBlur(image, (7, 7), 0)
plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(Gaussian, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Blurring')
cv2.imwrite("../images/exercise-images/gaussian_blur.jpg", Gaussian) 

# Median Blur
median = cv2.medianBlur(image, 5)
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))
plt.title('Median Blurring')
cv2.imwrite("../images/exercise-images/median_blur.jpg", median)  

# Bilateral Blur
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(bilateral, cv2.COLOR_BGR2RGB))
plt.title('Bilateral Blurring')
cv2.imwrite("../images/exercise-images/bilateral_blur.jpg", bilateral) 

# Show the plots
plt.tight_layout()
plt.show()
```
