# Image Cropping

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread("../images/exercise-images/sst.jpg", 0)

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

cropped_img = img[450:700, 200:500]

# Plot the cropped image
plt.subplot(1, 2, 2)
plt.title("Cropped Image")
plt.imshow(cropped_img)

# Save the cropped image
cv.imwrite("../images/exercise-images/cropped_out.jpg", cropped_img)

plt.show()
```

# Image Cropping with User Input

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

def crop_image_with_user_input(img):
    # Display the original image with dimensions
    plt.imshow(img, cmap='gray')
    plt.title(f"Original Image (Dimensions: {img.shape[1]}x{img.shape[0]})")
    plt.xlabel("Width (pixels)")
    plt.ylabel("Height (pixels)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Prompt user for cropping dimensions
    print("Please enter the cropping dimensions (x,y,w,h):")
    print("x = starting x-coordinate")
    print("y = starting y-coordinate")
    print("w = width")
    print("h = height")
    x, y, w, h = map(int, input("Enter values separated by commas (e.g., 100,100,200,200): ").split(','))

    # Crop the image
    cropped_img = img[y:y+h, x:x+w]

    # Display the cropped image
    plt.imshow(cropped_img, cmap='gray')
    plt.title("Cropped Image")
    plt.xlabel("Width (pixels)")
    plt.ylabel("Height (pixels)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Load the image
img = cv.imread("../images/exercise-images/sst.jpg", 0) # 0 for grayscale image

# Call the function to crop the image with user input
crop_image_with_user_input(img)
```
